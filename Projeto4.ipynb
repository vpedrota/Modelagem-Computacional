```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def simulacao(N = 500, I = 1, R = 0, r = 0.65, a = 0.1, t = 150):
    # N é a população total
    suscetivel = N - I - R #populaçao suscetivel
    
    step = 0.5 #passo da simulação
    s_array = [suscetivel]
    i_array = [I]
    r_array = [R]

    for i in np.arange(0,t,step):  
        dSdt = (-r * suscetivel * I / N) * step
        dIdt = (r * suscetivel * I / N - a * I) * step
        dRdt = (a * I) * step
        suscetivel += dSdt
        I += dIdt 
        R += dRdt

        s_array.append(suscetivel)
        i_array.append(I)
        r_array.append(R)
    
    plt.plot( s_array,  color='green', label='S')
    plt.plot(i_array, color='red', label='I')
    plt.plot( r_array, color='blue', label='R')
    plt.legend()
    plt.show()
    print("Número total de infectados: {:.0f}".format(R))
    print("{:.0f}% da população infectada".format(100*R/500))
```

### 1. Modelo matemático




        As seguintes equações descrevem o modelo matemático para essa simulação que utiliza o modelo SIR:        
$$
    \frac{dS}{dt} = -rSI;
$$

$$
    \frac{dI}{dt} = rSI - aI;
$$

$$
    \frac{dR}{dt} = aI;
$$



### 2. Parâmetros de simulação
\
$N$: Numero total da população
\
\
$S$: Quantidade de indivíduos suscetíveis a infecção da doença
\
\
$I$: Numero de infectados
\
\
$R$: Numero de recuperados
\
\
$t$: Tempo da simulação
\
\
$r$: taxa de contágio
\
\
$a$: taxa de recuperação

Nessa simulação temos um cenário em que não são tomadas medidas de prevenção ao COVID-19, então o que ocorre é uma curva acentuada que resulta na infecção de toda população num curto período de tempo, o que ocasiona na lotação dos leitos de hospitais.

```python
simulacao()
```
